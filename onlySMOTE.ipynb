```python
all_K_feature = pd.concat([K_pos_feature, K_neg_feature], ignore_index=True)
```

```python
import numpy as np
```

```python
K_label = np.zeros(len(all_K_feature), dtype = int)
```

```python
K_label[:618] = 1
```

```python
from collections import Counter
```

```python
from imblearn.over_sampling import BorderlineSMOTE

sm = BorderlineSMOTE(sampling_strategy= {1: 618 * 10}, random_state=42)
X_res, y_res = sm.fit_resample(all_K_feature, K_label)
```
